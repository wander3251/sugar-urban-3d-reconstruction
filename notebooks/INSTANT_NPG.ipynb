```python
import os
import lpips
import torch
from PIL import Image
import torchvision.transforms as transforms

loss_fn = lpips.LPIPS(net='alex').cuda()

transform = transforms.Compose([
    transforms.ToTensor()
])

gt_dir = "/content/data/artur/images"
pred_dir = "/content/renders"

files = sorted(os.listdir(gt_dir))

scores = []

for f in files:
    gt = Image.open(os.path.join(gt_dir, f)).convert("RGB")
    pred = Image.open(os.path.join(pred_dir, f)).convert("RGB")

    gt = transform(gt).unsqueeze(0).cuda()*2-1
    pred = transform(pred).unsqueeze(0).cuda()*2-1

    with torch.no_grad():
        score = loss_fn(gt, pred).item()

    scores.append(score)

print("LPIPS médio:", sum(scores)/len(scores))
```
