```python
from pathlib import Path

files = [
    "/content/SuGaR/sugar_extractors/refined_mesh.py",
    "/content/SuGaR/sugar_extractors/coarse_mesh.py",
    "/content/SuGaR/metrics.py",
    "/content/SuGaR/sugar_scene/sugar_model.py",
    "/content/SuGaR/gaussian_splatting/train.py",
]

for f in files:
    p = Path(f)
    txt = p.read_text()

    txt = txt.replace(
        "torch.load(",
        "torch.load("
    )

    txt = txt.replace(
        "map_location=",
        "weights_only=False, map_location="
    )

    p.write_text(txt)

print("Arquivos corrigidos.")
```
